```python
import pandas as pd
xls = pd.ExcelFile('Excel_test_dlouhy.xlsx')
dodavky = pd.read_excel(xls, 'dodavky')
dodavky
```

# Počet dodávek od dodavatele atd.

```python
dodavky.groupby(['ID_dodavatele']).count()
dodavky.groupby(['ID_dodavatele']).sum()
dodavky.groupby(['ID_dodavatele','ID_komponenty']).sum().sort_values(['Mnozstvi'],ascending=False)['Mnozstvi']
```

# Práce s časem (zpoždění)

```python
dodavky['doba_doruceni']=dodavky['Datum_dodani']-dodavky['Datum_objednani']
dodavky['doba']=dodavky['doba_doruceni'].astype(str)
dodavky['doba']=dodavky['doba'].str[0:2].astype(int)
dodavky.groupby(['ID_dodavatele']).mean()['doba']
```

# Standartní deviace zpoždění

```python
import numpy as np
dodavky['zpozdeni']=dodavky['doba']-7
dodavky.loc[~(dodavky['zpozdeni']>0),'zpozdeni']=np.nan
dodavky.groupby(['ID_dodavatele']).mean()['zpozdeni']
deviace=dodavky['zpozdeni']
deviace.std(skipna=True)
dodavky_dod=dodavky.loc[dodavky['ID_dodavatele']=='999DJT8']
deviace_dod=dodavky_dod['zpozdeni']
deviace_dod.std(skipna=True)
```
