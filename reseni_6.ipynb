```python
import pandas as pd
xls = pd.ExcelFile('Excel_test_dlouhy.xlsx')
dodavky = pd.read_excel(xls, 'dodavky')
```

```python
dodavky['Datum_objednani']=pd.to_datetime(dodavky['Datum_objednani'])
dodavky['Datum_dodani']=pd.to_datetime(dodavky['Datum_dodani'])
```

# Zpoždění?(poslední sloupec)

```python
dodavky['doba_doruceni']=dodavky['Datum_dodani']-dodavky['Datum_objednani']
dodavky['zpozdeni']=dodavky['doba_doruceni'].apply(lambda x: x.value)
dodavky['zpozdeni']=dodavky.loc[(dodavky['zpozdeni']>604800000000000)]
dodavky['zpozdeni']=dodavky['zpozdeni'].notnull().astype('int')
dodavky.head(20)
```
