```python
import pandas as pd
xls = pd.ExcelFile('Excel_test_dlouhy.xlsx')
vyroba = pd.read_csv('Excel_test_dlouhy.csv', sep = ';')
sklad = pd.read_excel(xls, 'sklad_08_2018')
z_ceho = pd.read_excel(xls, 'matice_vyroby')
vyroba_kopie=vyroba.copy()
```

# Ležák = v závodě se přes 2 měsíce z komponentu nic nevyrobilo

# 3 datasety pro každý závod

```python
vyroba['Datum']=pd.to_datetime(vyroba['Datum'])
vybrane_mesice=vyroba.loc[(vyroba['Datum']>='2018-08-01') & (vyroba['Datum']<'2018-09-30')]
vybrane_mesice
```

```python
plzen=vybrane_mesice.loc[vybrane_mesice['ID_zavodu']=='ZP10']
prerov=vybrane_mesice.loc[vybrane_mesice['ID_zavodu']=='ZP20']
ostrava=vybrane_mesice.loc[vybrane_mesice['ID_zavodu']=='ZP30']
prerov
```

# Komponenty v závodech (k 1.8.2018)

```python
import numpy as np
conditions = [
    (sklad['ID_skladu'] == 'SK01') | (sklad['ID_skladu'] == 'SK11')| (sklad['ID_skladu'] == 'SK86A'),
    (sklad['ID_skladu'] == 'SK56') | (sklad['ID_skladu'] == 'SK97'),
    (sklad['ID_skladu'] == 'SK00') | (sklad['ID_skladu'] == 'SK02')]
values=['ZP10','ZP20','ZP30']
sklad['ID_zavodu']=np.select(conditions,values)
sklad_plzen=sklad.loc[sklad['ID_zavodu']=='ZP10']
sklad_prerov=sklad.loc[sklad['ID_zavodu']=='ZP20']
sklad_ostrava=sklad.loc[sklad['ID_zavodu']=='ZP30']
sklad_ostrava
```

```python
komp={}
for x in z_ceho['ID_komponenty']:
    komp[x]=0
komp_plzen=komp.copy()
komp_ostrava=komp.copy()
komp_prerov=komp.copy()
```

# Kolik komponentů se využilo za 2 měsíce v každém závodu

```python
for index,komp in enumerate(plzen['ID_produktu']):
    for dil in z_ceho.loc[z_ceho['ID_produktu']==komp]['ID_komponenty']:
        komp_plzen[dil]+=int(z_ceho.loc[(z_ceho['ID_komponenty']==dil) & (z_ceho['ID_produktu']==komp)]['Mnozstvi'])*(int(plzen.iloc[index,3]))

for index,komp in enumerate(prerov['ID_produktu']):
    for dil in z_ceho.loc[z_ceho['ID_produktu']==komp]['ID_komponenty']:
        komp_prerov[dil]+=int(z_ceho.loc[(z_ceho['ID_komponenty']==dil) & (z_ceho['ID_produktu']==komp)]['Mnozstvi'])*(int(prerov.iloc[index,3]))

for index,komp in enumerate(ostrava['ID_produktu']):
    for dil in z_ceho.loc[z_ceho['ID_produktu']==komp]['ID_komponenty']:
        komp_ostrava[dil]+=int(z_ceho.loc[(z_ceho['ID_komponenty']==dil) & (z_ceho['ID_produktu']==komp)]['Mnozstvi'])*(int(ostrava.iloc[index,3]))
komp_plzen
```

# Zboží co jsou ležáci vůči lokalitám (pokud komponent > 0 tzn. zůstává ve skladu) (platí pouze pro KO---)

```python
for index,x in enumerate(sklad_plzen['ID_komponenty']):
    if x in komp_plzen:
        sklad_plzen.iloc[index,2]=sklad_plzen.iloc[index,2]-komp_plzen[x]
        
for index,x in enumerate(sklad_prerov['ID_komponenty']):
    if x in komp_prerov:
        sklad_prerov.iloc[index,2]=sklad_prerov.iloc[index,2]-komp_prerov[x]

for index,x in enumerate(sklad_ostrava['ID_komponenty']):
    if x in komp_ostrava:
        sklad_ostrava.iloc[index,2]=sklad_ostrava.iloc[index,2]-komp_ostrava[x]
sklad_prerov
```
