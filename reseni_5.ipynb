```python
import pandas as pd
xls = pd.ExcelFile('Excel_test_dlouhy.xlsx')
vyroba = pd.read_csv('Excel_test_dlouhy.csv', sep = ';')
z_ceho = pd.read_excel(xls, 'matice_vyroby')
vyroba_kopie=vyroba.copy()
vyroba
```

```python
vyroba['Datum']=pd.to_datetime(vyroba['Datum'])
posledni_mesic=vyroba.loc[(vyroba['Datum']>='2018-12-01')]
posledni_mesic
```

# List využívaných komponentů

```python
list_komp={}
z_ceho
for comp in z_ceho['ID_komponenty']:
    list_komp[comp]=0
```

# spotřeba komponent za měsíc (vyrobeno produktů * co je k výrobě potřeba)

```python
for index,komp in enumerate(posledni_mesic['ID_produktu']):
    for dil in z_ceho.loc[z_ceho['ID_produktu']==komp]['ID_komponenty']:
        list_komp[dil]+=int(z_ceho.loc[(z_ceho['ID_komponenty']==dil) & (z_ceho['ID_produktu']==komp)]['Mnozstvi'])*(int(posledni_mesic.iloc[index,3]))
list_komp
```
