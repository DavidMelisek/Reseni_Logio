```python
import pandas as pd
xls = pd.ExcelFile('Excel_test_dlouhy.xlsx')
komponenty = pd.read_excel(xls, 'matice_vyroby')
ceny = pd.read_excel(xls, 'ciselniky')
komponenty_kopie=komponenty.copy()
komponenty_kopie
```

```python
ceny_komponentu=ceny.iloc[:,9:11]
jmena_cols=ceny_komponentu.iloc[0]
ceny_komponentu=ceny_komponentu[1:]
ceny_komponentu.columns=jmena_cols
ceny_komponentu.reset_index()
ceny_komponentu
```

# Přiřazení ceny ke komponentům

```python
for index, komp in enumerate(komponenty_kopie['ID_komponenty']):
    komponenty_kopie['ceny'][index] = int(ceny_komponentu.loc[ceny_komponentu['ID_komponenty']==komp]['Porizovaci_cena'])
komponenty_kopie.head(10)
```

# Seřazení produktů podle celkové ceny všech komponentů

```python
komponenty_kopie['celk_cena']=komponenty_kopie['ceny']*komponenty_kopie['Mnozstvi']
komponenty_kopie.groupby(['ID_produktu']).sum()['celk_cena']
```
