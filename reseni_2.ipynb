```python
import pandas as pd
xls = pd.ExcelFile('Excel_test_dlouhy.xlsx')
vyroba = pd.read_csv('Excel_test_dlouhy.csv', sep = ';')
ceny = pd.read_excel(xls, 'ciselniky')
vyroba_kopie=vyroba.copy()
vyroba
```

# Vyrobeno dle závodu

```python
vyroba.groupby(['ID_zavodu']).sum()['Mnozstvi']
```

```python
ceny_produktu=ceny.iloc[:,12:14]
jmena_cols=ceny_produktu.iloc[0]
ceny_produktu=ceny_produktu[1:]
ceny_produktu.columns=jmena_cols
ceny_produktu.reset_index()
ceny_produktu
```

# Přiřazení ceny k produktům + znásobení množstvím vyrobených

```python
for index, komp in enumerate(vyroba_kopie['ID_produktu']):
    vyroba_kopie['ceny'][index]= int(ceny_produktu.loc[ceny_produktu['ID_produktu']==komp]['Prodejni_cena'])
vyroba_kopie
```

```python
vyroba_kopie['celk_cena']=vyroba_kopie['ceny']*vyroba_kopie['Mnozstvi']
vyroba_kopie.groupby(['ID_zavodu']).sum()['celk_cena']
```
